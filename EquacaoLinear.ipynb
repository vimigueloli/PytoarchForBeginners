Equação linear no contexto de separação

```python
## A equação da reta
import numpy as np

## variaveis
a = 3
b = -2
c = 1

## vetor de valores X - 50 valores entre -1 e 1
x = np.linspace(1,-1,50)

## equação:
## ax + by + c =0
# equação em função de y:
y = (-a*x-c)/b
```

```python
## Plotando um grafico
import matplotlib.pyplot as plt

## colocando a reta no grafico
plt.plot(x,y)

## colocano a grade no grafico
plt.grid(True)

## colocandoeixos x e y com a cor preta
# em 0 de -2 a +2, cor preta e com largura de 1 
plt.axvline(0,-2,2,color = 'k', linewidth = 1)
# em 0 de -2 a +2, cor preta e com largura de 1 
plt.axhline(0,-2,2,color = 'k', linewidth = 1)

```

```python
## colocando a plotagem do grafico em uma função

def plotGraph(a,b,c):

  x = np.linspace(1,-1,50)
  y = (-a*x-c)/b
  plt.plot(x,y)
  plt.grid(True)
  plt.axvline(0,-2,2,color = 'k', linewidth = 1)
  plt.axhline(0,-2,2,color = 'k', linewidth = 1)


plotGraph(3,-3,0)
```

```python
## Criando um ponto em cima da reta

## define o ponto
ponto = (0.5,0.5)

## mostra o grafico e o ponto
plotGraph(3,-3,0)
plt.plot(ponto[0],ponto[1], color='r', marker='o')
```

```python
## Criando um ponto acima da reta

## define o ponto
ponto = (0.5,0.5)
ponto1 = (0.25,0.75)

## mostra o grafico e o ponto
plotGraph(3,-3,0)
plt.plot(ponto[0],ponto[1], color='b', marker='o')
plt.plot(ponto1[0],ponto1[1], color='r', marker='o')
```

```python
## Criando um ponto abaixo da reta

## define o ponto
ponto = (0.5,0.5)
ponto1 = (0.25,0.75)
ponto2 = (0.75,0.25)

## mostra o grafico e o ponto
plotGraph(3,-3,0)
plt.plot(ponto[0],ponto[1], color='b', marker='o')
plt.plot(ponto1[0],ponto1[1], color='b', marker='o')
plt.plot(ponto2[0],ponto2[1], color='r', marker='o')
```

```python
## Retorno em relação a reta

ponto = (0.5,0.5)
ponto1 = (0.25,0.75)
ponto2 = (0.75,0.25)

a=-3
b=3
c=0

plotGraph(-3,3,0)
plt.plot(ponto[0],ponto[1], color='b', marker='o')
plt.plot(ponto1[0],ponto1[1], color='k', marker='o')
plt.plot(ponto2[0],ponto2[1], color='r', marker='o')

retorno = ponto[0]*a +b*ponto[1] + c
print('retorno do ponto sobre a reta')
print("%.2f" % retorno)
print('zero')
print('')

retorno = ponto1[0]*a +b*ponto1[1] + c
print('retorno do ponto acima a reta')
print("%.2f" % retorno)
print('positivo')
print('')

retorno = a*ponto2[0] +b*ponto2[1] + c
print('retorno do ponto abaixo a reta')
print("%.2f" % retorno)
print('negativo')
print('')
```
